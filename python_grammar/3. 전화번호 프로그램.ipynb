## 전화번호 프로그램 만들기

* 홍길동 010-123-4567
* 기능1 : 연락처 추가
* 기능2 : 연락처 전체 보기
* 기능3 : 검색, 이름을 입력받아서 전화번호 조회
* 기능4 : 수정, 이름을 입력받아서 전화번호 입력수정
* 기능5 : 삭제, 이름 입력받아서 삭제
* 기능6 : 프로그램 종료

```python
select = 0
phone = {}
while True:
    print('1번: 연락처 추가')
    print('2번: 연락처 전체')
    print('3번: 연락처 조회')
    print('4번: 연락처 수정')
    print('5번: 연락처 삭제')
    print('6번: 종료')
    select = int(input('\n원하시는 기능의 번호를 눌러주세요:'))
    print('='*40)

    
    if select==1:
        print('[연락처 추가]')
        
        name = input('이름을 입력해주세요:   ')
        num = input('연락처를 입력해주세요:  ')
        
        # phone[name] = num : 덮어쓰기 가능
        
        phone.setdefault(name, num)
        
        print('\n연락처가 추가되었습니다')
        print('='*40)
    
    elif select ==2:
        print('[연락처 전체]')
        
        if len(list(phone.keys())) != 0:
            for key, value in phone.items():
                print(f'{key} : {value}')
        else:
            print('저장된 연락처가 없습니다')
            
        print('='*40)
    
    elif select ==3:
        print('[연락처 조회]')

        
        search = input('찾으시는 이름을 입력해주세요:   ')
        print()
                
        phone.get(search, '\n해당하는 연락처가 없습니다')

        '''
        if search in list(phone.keys()):
            print(f'{search} : {phone[search]}')
        else:
            print('\n해당하는 연락처가 없습니다')
        '''
        
        print('='*40)
 
    elif select ==4:
        print('[연락처 수정]')
        
        search = input('수정하실 분의 이름을 입력해주세요:  ')
        
        if search in phone:
            update_num = input('수정할 연락처를 입력해주세요:  ')
            phone[search] = update_num
        else:
            print('\n해당하는 연락처가 없습니다')
        
        
        print('='*40)
    
    elif select ==5:
        print('[연락처 삭제]')
        
        delete = input('삭제하실 분의 이름을 입력해주세요:  ')
        
        if search in list(phone.keys()):
            del phone[delete]
        else:
            print('\n해당하는 연락처가 없습니다')
            
        print('='*40)
    
    elif select ==6:
        print('프로그램이 종료됩니다')
        break
    
    else:
        print('잘못된 번호입니다')
        print('='*40)
    
```

```python
contact = {}

while True:
    print('------ 전화번호부 프로그램 ------')
    print('1.추가   2.보기    3.검색   4.수정   5.삭제    9.종료')

    menu = int(input('메뉴를 선택해주세요(숫자 only)'))

    if menu == 1:
        print('연락처를 추가합니다.')
        new_name = input('이름 : ')
        new_tel = input('전화번호 : ')
        print(new_name, new_tel)
        # contact[new_name] = new_tel # 덮어쓰기
        contact.setdefault(new_name, new_tel)

    elif menu == 2:
        print('연락처를 조회합니다.')
        # print(contact)
        # print('='*10)
        for name, tel in contact.items():
            print(name, ":", tel)

    elif menu == 3:
        print('연락처를 검색합니다.')
        search_name = input('검색 이름: ')
        # print(contact[search_name])
        print(contact.get(search_name, '없는 연락처입니다.'))

    elif menu == 4:
        print('연락처를 수정합니다.')
        mod_name = input('수정 이름 : ') # 연락처에 있을 경우에만 새전화번호를 묻는 것이.
        if mod_name in contact:
            mod_tel = input('새 전화번호 : ')
            contact[mod_name] = mod_tel
        else :
            print('등록되지 않은 이름입니다.')

    elif menu == 5:
        print('연락처를 삭제합니다.')
        del_name = input('삭제 이름 : ')
        if del_name in contact:
            del contact[del_name]
        else :
            print('등록되지 않은 이름입니다.')

    elif menu == 9:
        print('프로그램을 종료합니다.') 
        break

    else :
        print('잘못된 입력입니다.')
```
